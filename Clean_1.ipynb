```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])
    
    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df['file_content'] = df['content'].apply(decode_method)

df.head()
```

```python
#Next we are cleaning the data by removing timestamp, non ascii characters,html tags etc

import re

def clean_text(file_content):
    # Remove HTML tags and other non-alphabetic characters
    cleaned_data = re.sub(r'<[^>]+>', '', file_content)
    cleaned_data = re.sub(r'[^a-zA-Z\s]', '', cleaned_data)
    
    # Remove non-ASCII characters
    cleaned_data = re.sub(r'[^\x00-\x7F]+', '', cleaned_data)
    
    # Remove timestamp patterns (if applicable)
    cleaned_data = re.sub(r'\d{2}:\d{2}:\d{2}\d{3} --> \d{2}:\d{2}:\d{2},\d{3}', '', cleaned_data)
    
    # Replace multiple spaces with a single space and convert to lowercase
    cleaned_data = re.sub(r'\s+', ' ', cleaned_data)
    cleaned_data = cleaned_data.lower().strip()
    
    return cleaned_data
```

```python
df['file_content'] = df['file_content'].apply(clean_text)
```

```python
def chunk_text(text, chunk_size=500, overlap=50):
    chunks = []
    words = text.split()
    for start in range(0, len(words), chunk_size - overlap):
        chunk = ' '.join(words[start:start + chunk_size])
        chunks.append(chunk)
    return chunks
```

```python
df['chunked_text'] = df['file_content'].apply(lambda x: chunk_text(x))
```
